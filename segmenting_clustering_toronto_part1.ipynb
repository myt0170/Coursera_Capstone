## Part 1: Transform the data on Wikipedia page into pandas dataframe

Import libraries

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

Scrap the wiki page by using BeautifulSoup

```python
url = 'https://en.wikipedia.org/w/index.php?title=List_of_postal_codes_of_Canada:_M&oldid=1011037969'
wiki_html = requests.get(url).text
soup = BeautifulSoup(wiki_html, 'html.parser')
```

Convert content of PostalCode HTML table as three lists of data

```python
# create three lists to store table data
postalCodeList = []
boroughList = []
neighborhoodList = []
```

```python
# add the data into the respective lists
for row in soup.find('table').find_all('tr'):
    cells = row.find_all('td')
    if(len(cells) > 0):
        postalCodeList.append(cells[0].text.rstrip('\n'))
        boroughList.append(cells[1].text.rstrip('\n'))
        neighborhoodList.append(cells[2].text.rstrip('\n'))
```

Create the dataframe

```python
# create a new DataFrame from the three lists
toronto_df = pd.DataFrame({"PostalCode": postalCodeList,
                           "Borough": boroughList,
                           "Neighborhood": neighborhoodList})

toronto_df
```

```python
# drop cells with a borough that is Not assigned
toronto_df_dropna = toronto_df[toronto_df.Borough != "Not assigned"].reset_index(drop=True)
toronto_df_dropna
```

Create a dataframe like the one shown in the instructions

```python
# create a new test dataframe
column_names = ["PostalCode", "Borough", "Neighborhood"]
instruc_df = pd.DataFrame(columns=column_names)

instruc_list = ["M5G", "M2H", "M4B", "M1J", "M4G", "M4M", "M1R", "M9V", "M9L", "M5V", "M1B", "M5A"]

for postcode in instruc_list:
    instruc_df = instruc_df.append(toronto_df[toronto_df["PostalCode"]==postcode], ignore_index=True)
    
instruc_df
```

Save the dataframe to csv for future tasks

```python
toronto_df_dropna.to_csv('toronto_postal_codes.csv', index=False)
```
